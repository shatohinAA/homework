```python
c = np.random.random(1000)
start = time.time()
a = np.array([0.12, 0.3, 0.167, 0.24, 0.31, 0.54, 0.111, 0.02, 0.001, 0.2])
b = a.cumsum()/a.cumsum()[-1]
def rng(i):
    if c[i]<=b[0]:
        return 0
    else:
        for j in range(1,10):
            if b[j-1]<c[i] and c[i]<=b[j]:
                return j
d = [rng(i) for i in range(1000)]
finish = time.time()
print(str(finish-start) + ' seconds')
```

```python
plt.hist(d)
plt.show()
```

Task 3

Разбрасывая таким образом точки, мы получаем равномерное распределение под графиком плотности изначальной функции, но он покрывает график нужной нам плотности, поэтому получается равномерное распределение и под графиком этой плотности.

```python
from scipy import stats
dist = stats.cauchy()
a = np.array([0]*1000, dtype=float)
i = 0
while i<1000:
    s=(np.random.random()-0.5)*8
    if np.random.random()<dist.pdf(s):
        a[i] = s
        i+=1
x = np.linspace(-4, 4, 100)
plt.hist(a, bins=50, density=True, label='Выборочная плотность')
plt.plot(x, dist.pdf(x), label='Точная плотность')
plt.legend()
plt.show()
```

```python
n = 100
def wandering(n):
    v = np.ones(2*n)
    mask1 = np.random.choice(np.arange(n),int(n/2),replace=False)
    mask2 = np.random.choice(np.arange(n,2*n),int(n/2),replace=False)
    v[mask1] = -1
    v[mask2] = -1
    v = v.reshape(2,n)
    A = np.array([1/np.sqrt(2),1/np.sqrt(2),-1/np.sqrt(2),1/np.sqrt(2)]).reshape(2,2)
    v = np.matmul(A,v).T/np.sqrt(2)
    v = v.cumsum(axis=0)
    v = np.vstack(([0,0], v))
    return v
v = wandering(n)
print('Координаты:\n')
print(v)
```

Task 7

Существует распределение хи-квадрат (распределение Пирсона), которое представляет собой распределение суммы квадратов k независимых нормальных случайных величин. И в случае, когда k = 2, это распределение является экспоненциальным.

Это означает, что если у точки в прямоугольной системе координат будут случайные координаты X и Y, распределенные нормально, то после перевода этих координат в полярную систему (r, θ) квадрат радиуса (расстояния от начала координат до точки) будет распределен по экспоненциальному закону, так как квадрат радиуса — это сумма квадратов координат (по закону Пифагора).

Так как плотность распределения равноценна во всех направлениях, угол θ будет иметь равномерное распределение в диапазоне от 0 до 2π. 
Для получения r и θ нужно сгенерировать две равномерно распределенные на отрезке (0, 1) случайные величины (назовем их u и v), распределение одной из которых (допустим v) необходимо преобразовать в экспоненциальное для получения радиуса. Проводя преобразования над функцией экспоненциального распределения, найдём радиус: r = sqrt(-2*log(v)).

Угол получим, растянув единичный отрезок до 2π: theta = 2*pi*u.
Теперь подставим r и θ в формулы перехода к полярной системе координат и получим искомые 2 независимые стандартные нормальные величины.

```python
n = 1000
u1, u2 = np.random.rand(2, n)
r = np.sqrt(-2 * np.log(u1))
theta = 2 * np.pi * u2
x = r * np.cos(theta)
y = r * np.sin(theta)
fig, ax = plt.subplots(1, 3, figsize=(15, 5))
ax[0].hist2d(x, y, bins=30)
ax[0].axis('equal')
ax[1].hist(x, bins=30)
ax[2].hist(y, bins=30)
ax[0].set_title("2d histogram")
ax[1].set_title("Histogram of x")
ax[2].set_title("Histogram of y")
plt.show()
```

```python
n1 = np.array([0]*1000, dtype=float)
n2 = np.array([0]*1000, dtype=float)
i=0
while i<1000:
    x = 2*np.random.random()-1
    y = 2*np.random.random()-1
    s = np.square(x)+np.square(y)
    if s>0 and s<=1:
        n1[i] = x*np.sqrt(-2*np.log(s)/s)
        n2[i] = y*np.sqrt(-2*np.log(s)/s)
        i+=1
A = np.array([1/2,np.sqrt(7)/2,2,0]).reshape(2,2)
v = np.vstack((n1, n2))
v = np.matmul(A,v).T
v[:,0] = v[:,0] + 3
v[:,1] = v[:,1] + 6
fig, ax = plt.subplots(1,1,figsize=(6, 8))
ax.hist2d(v[:,0], v[:,1], bins=30)
ax.axis('equal')
ax.set_title("2d histogram")
plt.show()
```
