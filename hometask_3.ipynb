```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
url = 'http://scipy-lectures.org/_downloads/populations.txt'
data = np.loadtxt(url, delimiter='\t', dtype=float, skiprows=1)
print(data)
```

```python
# Task 1
plt.plot(data[:, 0], data[:, 1], label='hare')
plt.plot(data[:, 0], data[:, 2], label='lynx')
plt.plot(data[:, 0], data[:, 3], label='carrot')
plt.xticks(np.arange(data[0][0],data[-1][0]+1,4))
plt.xlabel('Years', fontsize=14)
plt.ylabel('Population')
plt.legend()
plt.show()
```

```python
plt.bar(data[:, 0], data[:, 1]/(data[:, 1]+data[:, 2]), label='hare')
plt.bar(data[:, 0], data[:, 2]/(data[:, 1]+data[:, 2]), bottom=data[:, 1]/(data[:, 1]+data[:, 2]),label='lynx')
plt.bar(data[:, 0], data[:, 3]/(data[:, 1]+data[:, 2]), bottom=(data[:, 1]+data[:, 2])/(data[:, 1]+data[:, 2]),label='carrot')
plt.xticks(np.arange(data[0][0],data[-1][0]+1,4))
plt.xlabel('Years', fontsize=14)
plt.ylabel('Ratio of populations')
plt.legend()
plt.show()
```

```python
print('The mean of hare population:')
print(np.mean(data[:, 1]))
print('The std of hare population:')
print(np.std(data[:, 1]))
print('The mean of lynx population:')
print(np.mean(data[:, 2]))
print('The std of lynx population:')
print(np.std(data[:, 2]))
print('The mean of carrot population:')
print(np.mean(data[:, 3]))
print('The std of carrot population:')
print(np.std(data[:, 3]))
```

```python
plt.subplot(131)
plt.hist(data[:, 1],color='y', label='hare')
plt.vlines(np.mean(data[:, 1]),0,6)
plt.ylabel('Number of times')
plt.legend()
plt.subplot(132)
plt.hist(data[:, 2],color='r',label='lynx')
plt.vlines(np.mean(data[:, 2]),0,7)
plt.xlabel('Population', fontsize=14)
plt.legend()
plt.subplot(133)
plt.hist(data[:, 3], label='carrot')
plt.yticks(np.arange(0,5,1))
plt.vlines(np.mean(data[:, 3]),0,4)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
print('The largest population of hares was in '+str(int(data[np.argmax(data[:, 1])][0])))
print('The largest population of lynxes was in '+str(int(data[np.argmax(data[:, 2])][0])))
print('The largest population of carrots was in '+str(int(data[np.argmax(data[:, 3])][0])))
```

```python
# Задание про PCA для картинок 8 на 8
from sklearn.datasets import load_digits
digits = load_digits()
digits.data.shape
```

```python
images = digits.data.reshape(-1, 8, 8)
```

```python
i = 985
plt.imshow(images[i], cmap='gray')
plt.title('This is {}'.format(digits.target[i]))
plt.show()
```

```python
from sklearn.decomposition import PCA

pca = PCA()

X_transformed = pca.fit_transform(digits.data.astype(float))
X_transformed.shape
```

```python
# Task 2
y = [np.sum([pca.explained_variance_[i] for i in range(j)]) for j in range(1,65)]
plt.plot(np.arange(1,65), y)
plt.xlabel('Number of principal components')
plt.ylabel('Total explained variance')
plt.show()
```

```python
# Task 3
print('A number of principal components for 99%:')
print(np.where(y > 0.99*y[63])[0][0]+1)
print('A number of principal components for 95%:')
print(np.where(y > 0.95*y[63])[0][0]+1)
print('A number of principal components for 90%:')
print(np.where(y > 0.90*y[63])[0][0]+1)
print('A number of principal components for 85%:')
print(np.where(y > 0.85*y[63])[0][0]+1)
```

```python
# Task 4
i = 540
plt.imshow(digits.images[i], cmap='gray')
plt.title('Original')
plt.xlabel('64 components')
plt.show()
```

```python
pca = PCA(n_components=41)
X_transformed_1 = pca.fit_transform(digits.data.astype(float))
X_restored_1 = pca.inverse_transform(X_transformed_1)

plt.imshow(X_restored_1[i].reshape(8, 8), cmap='gray')
plt.title('99% of explained variance')
plt.xlabel('41 components')
plt.show()
```

```python
pca = PCA(n_components=29)
X_transformed_2 = pca.fit_transform(digits.data.astype(float))
X_restored_2 = pca.inverse_transform(X_transformed_2)

plt.imshow(X_restored_2[i].reshape(8, 8), cmap='gray')
plt.title('95% of explained variance')
plt.xlabel('29 components')
plt.show()
```

```python
pca = PCA(n_components=21)
X_transformed_3 = pca.fit_transform(digits.data.astype(float))
X_restored_3 = pca.inverse_transform(X_transformed_3)

plt.imshow(X_restored_1[i].reshape(8, 8), cmap='gray')
plt.title('90% of explained variance')
plt.xlabel('21 components')
plt.show()
```

```python
pca = PCA(n_components=17)
X_transformed_4 = pca.fit_transform(digits.data.astype(float))
X_restored_4 = pca.inverse_transform(X_transformed_4)

plt.imshow(X_restored_4[i].reshape(8, 8), cmap='gray')
plt.title('85% of explained variance')
plt.xlabel('17 components')
plt.show()
```
